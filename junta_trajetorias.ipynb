```python
arquivos = ["data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2011_2020.xlsx",
"data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2012_2021.xlsx",
# "data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2013_2021.xlsx",
"data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2013_2022.xlsx",
"data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2014_2023.xlsx",
"data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2015_2023.xlsx",
"data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2016_2023.xlsx",
"data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2017_2023.xlsx",
"data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2018_2023.xlsx",
"data/indicadores_trajetoria/indicadores_trajetoria_educacao_superior_2019_2023.xlsx"]
```

```python
import pandas as pd
```

```python
all_dfs = []
for file_name in arquivos:
    # Create the full file path
    # file_path = os.path.join(data_dir, file_name)
    
    # Read the Excel file with the same parameters you used
    df = pd.read_excel(file_name, header=[8])
    df = df.iloc[:-1]
    
    # Optional: Add a column to identify which file this data came from
    df['source_file'] = file_name
    
    # Append to our list of dataframes
    all_dfs.append(df)
    
    # Print confirmation (optional)
    print(f"Loaded {file_name}, shape: {df.shape}")

# Combine all dataframes into one
raw_df = pd.concat(all_dfs, ignore_index=True)
```
